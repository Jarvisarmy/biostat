```python
from collections import Counter
import json
```

```python
def cut(lst, length, overlap):
    ret = []
    inter = length-overlap
    for i in range(0,len(lst),inter):
        this_append = []
        for x in lst[i:i+length]:
            this_append.append(x)
            if (len(this_append)==length):
                ret.append(this_append)
    return ret
```

```python
cut("PPYTVVYFPVRGRCAALRMLLA",4,1)
```

```python
def get_train_dict(train_path, length, overlap):
    all_train = []
    for ll, line in enumerate(open(train_path,'r',encoding='utf-8')):
        line = line.strip()
        if 'X' in line:
            continue
        elif 'U' in line:
            continue
        elif 'Z' in line:
            continue
        elif 'B' in line:
            continue
        elif 'O' in line:
            continue
        words = cut(line, length,overlap)
        for word in words:
            all_train.append(''.join(word))
    res = Counter(all_train)
    train_dict = dict(res)
    return train_dict
```

```python
get_train_dict('data/train_36w',5,1)
```

```python
def get_test(test_path, length, overlap):
    res = []
    for ll, line in enumerate(open(test_path,'r',encoding='utf-8')):
        line = line.strip()
        pos = [str(l) for l in range(1,len(line)+1)]
        poss = cut(pos,length,overlap)
        lines = cut(line,length,overlap)
        temp_res = []
        for line, pos in zip(lines, poss):
            line = ''.join(line)
            pos = ' '.join(pos)
            temp_res.append([line, pos])
        res.append(temp_res)
    return res[0],res[1],res[2]
            
            
```

```python
get_test('data/test_loop',5,1)
```

```python
def get_test_dict(train_dict, eg5_test, eg7_test, eg7_test2):
    eg5_dict = {}
    eg7_dict1 = {}
    eg7_dict2 = {}
    for sets,dicts in zip((eg5_test, eg7_test, eg7_test2),(eg5_dict,eg7_dict1,eg7_dict2)):
        for s in sets:
            if s[0] not in train_dict.keys():
                dicts[str(s)] = 0
            else:
                dicts[str(s)] = train_dict[s[0]]
    return eg5_dict, eg7_dict1, eg7_dict2
```

```python
train_dict = get_train_dict('data/train_36w',5,1)
eg5_test, eg7_test1, eg7_test2 = get_test('data/test_loop',5,1)
get_test_dict(train_dict, eg5_test, eg7_test1, eg7_test2)
```

```python
def save_cipin(eg5_dict, eg7_dict1, eg7_dict2, length, overlap,train_dict):
    with open('result/dict_%s_%s.txt'%(length,overlap), 'w', encoding="utf-8")as f, open('result/train_dict_%s_%s.txt'%(length,overlap), 'w', encoding="utf-8")as f1:
        len1 = abs(len(eg7_dict2)-len(eg5_dict))
        len2 = abs(len(eg7_dict2)-len(eg7_dict1))
        eg5_dict = eg5_dict+['None']*len1
        eg7_dict1 = eg7_dict1+['None']*len2
        summ = sum(train_dict.values())
        for eg7_1, eg5, eg7_2 in zip(eg7_dict1,eg5_dict,eg7_dict2):
            f.write(json.dumps(summ) + '\n' + 'EG7_1:\t' + json.dumps(eg7_1) + '\t' + 'EG7_2:\t' + json.dumps(eg7_2) + '\t' + 'EG5:\t' + json.dumps(eg5) + '\n')
        print(json.dumps(train_dict))
        f1.write(json.dumps(train_dict)+'\n')
```

```python
train_dict = get_train_dict('data/train_36w',5,1)
eg5_test, eg7_test1, eg7_test2 = get_test('data/test_loop',5,1)
eg5_dict, eg7_dict1, eg7_dict2 = get_test_dict(train_dict, eg5_test, eg7_test1, eg7_test2)
eg5_dict = sorted(eg5_dict.items(), key=lambda x: x[1])
eg7_dict1 = sorted(eg7_dict1.items(), key=lambda x: x[1])
eg7_dict2 = sorted(eg7_dict2.items(), key=lambda x: x[1])
```

```python
save_cipin(eg5_dict, eg7_dict1, eg7_dict2,5,1,train_dict)
```
